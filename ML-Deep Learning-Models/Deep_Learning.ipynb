# What Is Deep Learning


- Deep Learning is a subfield of machine learning concerned with algorithms inspired by the structure and function of the brain called Artificial Neural Networks.

- Artificial neural networks (ANNs), usually simply called neural networks (NNs), are computing systems vaguely inspired by the biological neural networks that constitute animal brains.

- Similar to Neurons in the brain, the Artificial Neural network consists of Neurons which takes in input do some process inside them(i.e implement some function) and outputs the result. An arrangement of these neural networks in a vertical way is called a Layer and these layers are stacked together to form a Network with inter-connections between them and so-called as Artificial Neural Network.

- It can be for supervised, semi-supervised, or unsupervised.

- There are many architectures in deep learning making them efficient for different fields like Convolution Neural Networks for Images, Recurrent Neural Network and it's variants for Natural Language Processing, Deep Neaural networks for regression and classification.

# How Neural Networks work ?

### The process is carried in 2 phases
1. Forward Propagation
2. Backward Propagation
 
 
- **Forward propagation** (or forward pass) refers to the calculation and storage of intermediate variables (including outputs) for a neural network in order from the input layer to the output layer. In other words when an example from the training set is fed to the input layer, the processing/computation and storage of that computation of neurons on the given example in the forward direction( input => output).
- This computation value at the final layer i.e Output layer is used to calculate the loss function( A Loss Function is a mathematical function used to measure the difference between the given output by Neural network and the original output we should get. They vary from Mean Square error to LogLoss)
- **BackPropogation** refers to the method of calculating the gradient of neural network parameters. This method traverses the network in reverse order, from the output to the input layer, according to the chain rule from calculus. In short, think of it like feedback which helps the neural network to understand, how for each neuron the priority/weight should be assigned from the value obtained from loss function such that the model will be best fit for our problem.

# Convolution Neaural Networks
- These are the type of neural networks that are best suited for Image classification. Instead of using Fully connected neural networks where each pixel of given image is considered as a feature, we use features like edges in images.
- The more layers you add the more complex features can be learnt
for example the first layer learns simple features like horizontal,verical edges while last layer can identify face of a cat/dog
- It is done by using filters and applying convolution . Lets see what they are with simple example 
lets take a matrix and assume this as a matrix representing image where 1 is dark and 0 is lighter part<br>
1 1 1 0 0<br>
0 1 1 0 0<br>
0 0 1 1 1<br>
0 0 1 1 0<br>
0 1 1 0 0<br>
- Is convoluted with a 3*3 filter<br>
1 0 -1<br>
1 0 -1<br>
1 0 -1<br>
 
- When the above filter is used to convolute with given matrix of pixels we get<br>
2 1 2<br>
3 -1 2<br>
3 1 2<br>
 
## How it is computed 
- Take the filter to place it on top of matrix b/w (0,0) and (2,2) now do element-wise multiplication and add them to get the first entry in the resultant matrix, now move the filter towards the right by one element i.e 0,1 and 2,3 and do the same to get the second element and so on ...
 
## What it signifies
- lets take<br> 
1 1 1<br>
1 1 1<br>
1 1 1<br>
is convoluted with our filter which gives **zero** as output but when<br>
1 1 0<br>
1 1 0<br>
1 1 0<br>
is convoluted with our filter which gives **3** as output which signifies there is a **vertical edge**(1 represents darker pixel and 0 represents lighter part)
 
## Takeaway 
- The filter above is used to detect vertical edges, where as filter below is used to detect horizontal edges<br>
1 1 1<br>
0 0 0<br>
 -1 -1 -1<br>
- 1 0 -1 mean if image has a variation from higher to lower transition of pixel intensity(consider as black to white (1=>-1)). So if 1's 0's -1's are stacked in vertical position it means we are searching for vertical edges. If stacked horizontally it means we are searching for horizontal edges.
- So our convolution network learns about how to design a filter that fits very well for our training set to extract features like horizontal edges, vertical edges, slanted edges and so on.

# Recurrent Neural Network

- This architecture does the thing. It is like a loop i.e you feed everything to one layer which passes its output as input to itself every time a new feature is given as input, thus having the memory of previous feature input.

# General Tuning Parametes
- **Batch Size** - 
It represents nuber of images that we will send as input into our network at a time<br>
If we sent small batch the training time will be more

- **Number of Epochs** - Number of times you need to send entire training set.

- **Learning rate** - As the gradient provides slope at that point we add a multiplying factor to the slope to decrese the parameters to find optimal solution(Gradient decent)
<br>With Learning rate(Multiplying factor) less we move slowly which takes more training time<br>
with Learning rate(Multiplying factor) more we take large steps which may lead to crossing of global minimum

# Regularization
Regularisation is a technique used to reduce the errors by fitting the function appropriately on the given training set and avoid overfitting
- Early Stopping - If you dont see any considerable change in Validation accuracy then stop training.
- Drop out - On each epoch Drop some neurons randomly introducing regualrization effect
- Make it simple - If your model is overfitting it means your made a complex model than required. Try decreasing hidden layers, number of neurons per layer.

# Activation Functions
- Basically these are the computational units in a neuron, these are responsible for the learnong of a neuron
- These are nothing but a function that neuron do on theinput it got and sends the output to other neurons
- These are of Different types
1. Linear (returns what it gots)
1. Sigmoid, Logistic, TanH(non linear, Computationally expensive)
1. Relu(Rectified Linear function) and its varients (computationally cheap and introduces non-linearity)
```
f(x) = x if x>=0
f(x) = 0 if x<0
```
1.Softmax(Used on the final output layer of a
multiple classification problem, it converts outputs to
probabilities of each classification)

## which to choose

- For multiple classification, use softmax on the output layer
- If it is a binary class use Sigmoid on the output layer
- RNN’s do well with Tanh
- For everything else use ReLU and its varients

# Importing Libraries

````python
import tensorflow
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input
from tensorflow.keras.layers import Dense,Flatten,BatchNormalization,Dropout,Conv2D,MaxPooling2D,Activation,MaxPool2D
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow.keras.backend as K
````

#### Using ImageDataGenerator We can skip the labeling of each image
 
- How does our model know which is a positive example and which is a negative one?
 
- Just provide the Directory/Folder that contains Number of Folders equivalent to the number of classes your training set needs to be trained to, and images of that class in their respective Folder
 
- Here Train is the parent folder. It contains 2 subfolders IDC(Positive) and NONIDC(Normal) with the photos representing them

````python
Train = '/tmp/tmp/Train/'
data_gen = ImageDataGenerator(
    preprocessing_function = preprocess_input,#The preprocessing function does preprocessing before inputting images to resnet50 as specified in the research paper. see next cell to understand ResNet50
    validation_split = 0.2
)
# Train(80%) and Validation(20%)
train_data = data_gen.flow_from_directory(
    Train,#Path of directory
    target_size=(50,50),#Resolution of image you want to input into Neaural network
    class_mode = 'binary',#How many classes(Binary for 2 class classification)
    batch_size=2048,#How many images you want to process at once(This is an hyper parameter tweak then to play with them)
    subset='training'#It's a Training Set
)
validation_data = data_gen.flow_from_directory(
    Train,#Path of the directory
    target_size=(50,50),#Resolution of image you want to input into Neaural network
    class_mode = 'binary',#How many classes(Binary for 2 class classification)
    batch_size=256,#How many images you want to process at once
    subset='validation'#It's a Validation Subset 
)
````

````python
train_data.class_indices #Prints the class names assigned to the directories inside the Train Folder
````

## This term is worth noting **Transfer Learning**
- The above term is self-explanatory i.e we are transferring the learning done by others. In other words as training, neural networks are computationally expensive and take up to several days to train we are using someones else training results and apply them on ours so that we can 
1. Spend less time on training
1. Use their feature detector and apply it on ours
1. start with better slope and asymptote
 
#### In the below we are using ResNet50 a famous NeuralNet architecture and apply transfer learning on it.
- As we only need a binary classifier a single neuron will do the work but before we add that on top of ResNet we need to make the layers of ResNet nontrainable so that it saves computation and time and Flatten the output since ResNet outputs a matrix.
- Resnet are good for vanishing gradient problem
- **Vanishing Gradient** when we approch to a minima the slope will become zero hence vanishing gradient. What resnet does is insted of passing the output of the prev layer to next layer it combines it with the outputs of before layers so that even if current layers gradient tends to zero prev gradients help them.

````python
decay_lr = ReduceLROnPlateau(factor=0.01,patience=1,min_lr=0.000001,monitor='val_accuracy')
# Reduces the Learning rate if there is a no change in val_accuracy  
callbacks = [decay_lr]
# Callbacks are the call backs that happen after each epoch
````

````python
pretrained_model = ResNet50(include_top=False,weights='imagenet',input_shape=(50,50,3)) #Resnet50 with its weights loaded from training on ImageNet dataset 
#and setting the input shape to the shape of images we set in ImageDataGenerator 
for layer in pretrained_model.layers: #Making Layers Non Trainable 
  layer.trainable = False
x = pretrained_model.output #Taking the output
x = Flatten()(x)#Passing the output to Flatten Layer
x = Dense(64,activation='sigmoid')(x)#Passing the Layer a Dense layer i.e a layer with 64 Neuron
x = Dense(1,activation='sigmoid')(x)#Passing the Layer to single Neuron
model = Model(pretrained_model.input,x) #Defining the model which takes input as our image thatis fed intp pretrained_model and output as the singlr Neuron
model.compile(optimizer='adam',metrics=['accuracy'],loss='binary_crossentropy')#Compiling model
````

# Clear the session and run the model


````python
K.clear_session()
history = model.fit(x = train_data, epochs=10,validation_data = validation_data,callbacks=callbacks)#Run the model, Epochs is number of times you want to run the model on dataset
#(Its a hyperparameter tweak it to learn which is more suitable
````

## Wondering about the difference in training and validation accuracy.
It is because the model we selected is complicated, as we have fewer data to train. So instead of recognizing patterns, it starts to root learn training set data, this is otherwise called Overfitting, i.e. model does very nice on training data but performs less on validation/test data.
### some ways to overcome this
- Less complicated model
- Regularization
<br>Regularization is a technique to reduce the complexity of the model. It does so by adding a penalty term to the loss function.
- Early Stopping
<br>We stop training our models when the validation error starts to increase.
- Data Augmentation
<br>Data augmentation means the increasing size of the data that is increasing the number of images present in the dataset. Some of the popular image augmentation techniques are flipping, translation, rotation, scaling, changing brightness, adding noise etc.
- Dropouts
<br>Dropout is a regularization unlike the original it avoids Overfitting not by modifying cost but by changing the network itself. It randomly drops neurons from the neural network during training in each iteration.


# Plot the accuracy of Training and validation set

````python
acc = history.history['accuracy'] #Extracts the accuracy from logs
val_acc = history.history['val_accuracy']#Extracts the validation accuracy from logs
epochs = range(len(acc))#Number of epochs = length of accuracy since each epoch has one accuracy
plt.plot(epochs, acc, 'r') #Creating a plot and plotting acc
plt.plot(epochs, val_acc, 'b') #Plotting val_acc in same plot
plt.title('Training and validation accuracy')#Name of the plot
plt.xlabel("Epochs")#X axis label
plt.ylabel("Accuracy")#Y axis label
plt.legend(["Training Accuracy", "Validation Accuracy"])#Representation
plt.figure()
````
